```python
# Project 1: Rule-Based AI Chatbot
# DecodeLabs | AI Intern
# Bot Name: Nova

import random

def main():
    # =============================================
    # KNOWLEDGE BASE
    # =============================================
    responses = {
        # ---- Greetings ----
        "hello":     "Hey there! I'm Nova, your AI assistant. What can I do for you?",
        "hi":        "Hi! Nova at your service.",
        "hey":       "Hey! How's it going?",
        "good morning": "Good morning! Ready to crush some code?",
        "good evening": "Good evening! Hope your day was productive.",

        # ---- Well-being ----
        "how are you":     "Running at 100% efficiency! Thanks for asking. You?",
        "how are you doing": "All systems green! How about yourself?",
        "i am fine":       "Glad to hear that!",
        "i am good":       "Awesome. Let's keep the vibe going.",
        "i am sad":        "Sorry to hear that. Want to talk about it or hear a joke?",

        # ---- About Nova ----
        "what is your name": "I'm Nova — your rule-based chatbot, built with pure Python logic.",
        "who are you":    "I'm Nova, a deterministic AI. No neural networks here — just clean if-this-then-that logic. Ask me what I can do!",
        "what can you do": "I can chat, tell jokes, remember your name, and respond to a bunch of predefined intents. Still growing!",
        "who created you": "I was built by an AI intern at DecodeLabs as Project 1 — the foundation of AI engineering.",

        # ---- Jokes (with variety via random) ----
        "tell me a joke": "random_joke",
        "another joke":   "random_joke",
        "joke":           "random_joke",
        "make me laugh":  "random_joke",

        # ---- Gratitude ----
        "thanks":     "Anytime! That's why I'm here.",
        "thank you":  "You got it. Need anything else?",
        "thank you so much": "No problem at all!",

        # ---- Help / Commands ----
        "help":       "Here's what you can try: hello, how are you, tell me a joke, what is your name, my name is [your name], what can you do",
        "commands":   "Try: hello | how are you | tell me a joke | what is your name | my name is ___ | bye",
    }

    # ---- Joke bank ----
    jokes = [
        "Why do programmers prefer dark mode? Because light attracts bugs! 😄",
        "How many programmers does it take to change a lightbulb? None — that's a hardware problem!",
        "Why did the Python programmer wear glasses? Because they couldn't C#!",
        "What's a chatbot's favorite drink? Java!",
        "I told my computer I needed a break... now it won't stop sending me Kit-Kats.",
        "Why do AI engineers love dictionaries? Because they hate long if-elif ladders!",
    ]

    # ---- Memory ----
    user_name = None

    # =============================================
    # GREETING
    # =============================================
    print("=" * 50)
    print("🤖  Nova v1.0 | DecodeLabs Project 1")
    print("=" * 50)
    print("Nova: Hi! I'm Nova, your rule-based assistant.")
    print("Nova: Type 'help' to see what I can do, or 'bye' to exit.")
    print("-" * 50)

    # =============================================
    # INFINITE LOOP (The Heartbeat)
    # =============================================
    while True:
        raw_input = input("You: ")
        user_input = raw_input.lower().strip()

        # ---- Handle empty input ----
        if not user_input:
            print("Nova: ...(crickets) Try typing something!")
            continue

        # ---- EXIT STRATEGY ----
        if user_input in ['bye', 'exit', 'quit', 'goodbye']:
            if user_name:
                print(f"Nova: Goodbye, {user_name}! Come back anytime.")
            else:
                print("Nova: Goodbye! Have a great day.")
            break

        # ---- Name memory (pattern match) ----
        if user_input.startswith("my name is "):
            user_name = user_input[11:].strip().title()
            print(f"Nova: Nice to meet you, {user_name}! I'll remember that.")
            continue

        if user_input in ["what is my name", "do you remember me", "who am i"]:
            if user_name:
                print(f"Nova: You're {user_name}! I've got a memory like a dictionary — O(1) lookup.")
            else:
                print("Nova: You haven't told me your name yet. Try: my name is [your name]")
            continue

        # ---- Lookup + fallback + joke handler ----
        reply = responses.get(user_input, None)

        if reply == "random_joke":
            reply = random.choice(jokes)
        elif reply is None:
            # ---- UNKNOWN INPUT: Friendly fallback ----
            fallbacks = [
                "Hmm, I'm still learning. Try 'help' to see what I understand.",
                "I didn't catch that. Maybe try rephrasing? Or type 'help'.",
                "My rule-book doesn't have that one yet. Give 'help' a shot!",
                "That's beyond my programming for now. But I'm great at jokes — try 'tell me a joke'!",
            ]
            reply = random.choice(fallbacks)

        print(f"Nova: {reply}")

    print("=" * 50)
    print("Session ended.")

if __name__ == "__main__":
    main()
```
